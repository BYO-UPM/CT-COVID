```python
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Load the COCO format json file
with open('/media/my_ftp/BasesDeDatos_Torax_RX_CT/COVID19_CT/processed/object_detection/coco_annotations/coco_train.json', 'r') as f:
    data = json.load(f)

# Load the image
image_data = data['images'][0]  # Change this to select different images
img = Image.open(image_data['file_name'])

# Create figure and axes
fig, ax = plt.subplots(1)

# Display the image
ax.imshow(img)

# Draw bounding boxes
for ann in data['annotations']:
    if ann['image_id'] == image_data['id']:
        bbox = ann['bbox']
        rect = patches.Rectangle((bbox[0], bbox[1]), bbox[2], bbox[3], linewidth=1, edgecolor='r', facecolor='none')
        ax.add_patch(rect)

plt.show()
```
